# To Plot using sns.relplot, line plot, box plot on fmri data set

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset('fmri')
df.head()
```

RELATIONAL PLOT

```python
sns.relplot(x='timepoint',y='signal',hue='subject',style='event',size='region',data=df)
```

LINE PLOT

```python
sns.relplot(x='timepoint',y='signal',hue='subject',style='event',size='region',data=df,kind='line')
```

BOX PLOT

```python
sns.boxplot(x='timepoint',y='signal',hue='event',data=df)
```
